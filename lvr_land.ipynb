```python
import pandas as pd
```

```python
import glob
```

```python
source_csv = glob.glob('lvrland_crawler/lvr_src/*/*/*/*/*_lvr_land_*.csv')
print(len(source_csv))
```

```python
li = []
for s in source_csv:
    sp = s.split('/')
    year, season, city, trade_type = sp[2], sp[3], sp[4], sp[5]
    df = pd.read_csv(source_csv[0], header = 1)
    new_column = f'{year}_{season}_{city}_{trade_type}'
    df['df_name'] = new_column
    li.append(df)
```

```python
lvrland_df = pd.concat(li, axis=0 , ignore_index = True)
lvrland_df.shape
lvrland_df
```

```python
is_self_residence = lvrland_df[lvrland_df['main use']=='住家用']
is_self_residence['main use']
```

```python
is_self_residence_building = lvrland_df[lvrland_df['building state'].str.startswith('住宅大樓')]
is_self_residence_building['building state']
```
